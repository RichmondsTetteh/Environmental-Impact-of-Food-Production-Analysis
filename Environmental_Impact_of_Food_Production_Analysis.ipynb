# Data Cleaning

```python
# Make copies of original pieces of data
cleaned_data = food_production.copy()
```

Based on the assessment of the food_production dataset, the most crucial data cleaning step I recommend are listed below;
1. Column name
2. Missing values

For handling the missing data I recommend the Median Imputation because:

Robustness to Outliers: The median is less affected by extreme values (outliers) compared to the mean. If your numerical columns have outliers, using the median for imputation will provide a more representative central value and prevent the imputed values from being skewed.

Simplicity: Median imputation is relatively straightforward to implement.

Code:

```python
# Identify numerical columns with missing values
numerical_cols_with_missing = cleaned_data.select_dtypes(include=np.number).columns[cleaned_data.select_dtypes(include=np.number).isnull().any()]
print(numerical_cols_with_missing)
```

```python
# Handle missing values in numerical columns by imputing with the median
for col in numerical_cols_with_missing:
    cleaned_data[col] = cleaned_data[col].fillna(cleaned_data[col].median())
```

Test:

```python
# Verify that missing values have been handled
print("\nMissing values after median imputation:")
print(cleaned_data.isnull().sum())
```

Column name

```python
# Create a dictionary to map old column names to new ones
new_column_names = {}
for col in cleaned_data.columns:
    # Convert to lowercase and replace spaces and hyphens with underscores
    new_name = col.lower().replace(' ', '_').replace('-', '_')

    # Shorten some of the longer names based on common patterns
    if 'eutrophying_emissions_per_1000kcal_(gpo₄eq_per_1000kcal)' in new_name:
        new_name = new_name.replace('eutrophying_emissions_per_1000kcal_(gpo₄eq_per_1000kcal)', 'e_emissions_gpo₄eq_1000kcal')
    elif 'eutrophying_emissions_per_kilogram_(gpo₄eq_per_kilogram)' in new_name:
        new_name = new_name.replace('eutrophying_emissions_per_kilogram_(gpo₄eq_per_kilogram)', 'e_emissions_gpo₄eq_kg')
    elif 'eutrophying_emissions_per_100g_protein_(gpo₄eq_per_100_grams_protein)' in new_name:
        new_name = new_name.replace('eutrophying_emissions_per_100g_protein_(gpo₄eq_per_100_grams_protein)', 'e_emissions_gpo₄eq_100g_protein')
    elif 'freshwater_withdrawals_per_1000kcal_(liters_per_1000kcal)' in new_name:
        new_name = new_name.replace('freshwater_withdrawals_per_1000kcal_(liters_per_1000kcal)', 'fw_withdrawals_l_1000kcal')
    elif 'freshwater_withdrawals_per_100g_protein_(liters_per_100g_protein)' in new_name:
        new_name = new_name.replace('freshwater_withdrawals_per_100g_protein_(liters_per_100g_protein)', 'fw_withdrawals_l_100g_protein')
    elif 'freshwater_withdrawals_per_kilogram' in new_name:
        new_name = new_name.replace('freshwater_withdrawals_per_kilogram_(liters_per_kilogram)', 'fw_withdrawals_l_kg')
    elif 'greenhouse_gas_emissions_per_1000kcal_(kgco₂eq_per_1000kcal)' in new_name:
        new_name = new_name.replace('greenhouse_gas_emissions_per_1000kcal_(kgco₂eq_per_1000kcal)', 'ghg_emissions_kgco2_1000kcal')
    elif 'greenhouse_gas_emissions_per_100g_protein_(kgco₂eq_per_100g_protein)' in new_name:
        new_name = new_name.replace('greenhouse_gas_emissions_per_100g_protein_(kgco₂eq_per_100g_protein)', 'ghg_emissions_kgco2_100g_protein')
    elif 'land_use_per_1000kcal_(m²_per_1000kcal)' in new_name:
        new_name = new_name.replace('land_use_per_1000kcal_(m²_per_1000kcal)', 'land_use_m2_1000kcal')
    elif 'land_use_per_kilogram_(m²_per_kilogram)' in new_name:
        new_name = new_name.replace('land_use_per_kilogram_(m²_per_kilogram)', 'land_use_m2_kg')
    elif 'land_use_per_100g_protein_(m²_per_100g_protein)' in new_name:
        new_name = new_name.replace('land_use_per_100g_protein_(m²_per_100g_protein)', 'land_use_m2_100g_protein')
    elif 'scarcity_weighted_water_use_per_kilogram_(liters_per_kilogram)' in new_name:
        new_name = new_name.replace('scarcity_weighted_water_use_per_kilogram_(liters_per_kilogram)', 'sw_water_use_l_kg')
    elif 'scarcity_weighted_water_use_per_100g_protein_(liters_per_100g_protein)' in new_name:
        new_name = new_name.replace('scarcity_weighted_water_use_per_100g_protein_(liters_per_100g_protein)', 'sw_water_use_l_100g_protein')
    elif 'scarcity_weighted_water_use_per_1000kcal_(liters_per_1000_kilocalories)' in new_name:
        new_name = new_name.replace('scarcity_weighted_water_use_per_1000kcal_(liters_per_1000_kilocalories)', 'sw_water_use_l_1000kcal')

    new_column_names[col] = new_name

# Rename the columns
cleaned_data.rename(columns=new_column_names, inplace=True)
```

# 3. Which food category/product has the highest individual Environmental Impact?

```python
# Set plot style
sns.set_style('whitegrid')

# Select relevant columns for plotting
environmental_impact_cols = ['land_use_change', 'animal_feed', 'farm', 'processing', 'transport', 'packging', 'retail']

# Melt the DataFrame to have categories as x-axis, impact type as hue
melted_data = cleaned_data.melt(id_vars=['food_product', 'category'],
                                value_vars=environmental_impact_cols,
                                var_name='impact_type',
                                value_name='value')

# Group by category and impact type to get the sum
grouped_impact = melted_data.groupby(['category', 'impact_type'])['value'].sum().reset_index()

# Pivot the data to create a stacked bar chart
pivot_impact = grouped_impact.pivot(index='category', columns='impact_type', values='value').fillna(0)

# Calculate total impact per category for sorting
category_total_impact = pivot_impact.sum(axis=1).sort_values(ascending=False)

# Sort the pivoted data by total impact
pivot_impact_sorted = pivot_impact.loc[category_total_impact.index]

# Create a stacked bar plot
plt.figure(figsize=(16, 8))
ax = pivot_impact_sorted.plot(kind='bar', stacked=True, colormap='viridis', ax=plt.gca())

plt.xticks(rotation=45, ha='right')
plt.xlabel('Food Category')
plt.ylabel('Total Impact')
plt.title('Total Environmental Impact per Food Category by Impact Type')
plt.tight_layout()

# Move the legend outside the plot
ax.legend(title='Impact Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add data labels to each segment of the stacked bars
# Define a threshold for displaying labels
label_threshold = 4.0 # Adjust this value as needed

for container in ax.containers:
    labels = [f'{w:.1f}' if w > label_threshold else '' for w in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='center', color='black')

plt.show()
```
